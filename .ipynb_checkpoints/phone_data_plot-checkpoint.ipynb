```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf as pdf

# Output text as editable PDF text objects instead of shapes
matplotlib.rcParams['pdf.fonttype'] = 42
# Force PDF to respect zorder by supressing image compositing
matplotlib.rcParams['image.composite_image'] = False

%matplotlib inline
```

```python
# Clean the original dataset and remove identifying information

full_phone = pd.read_csv('./phone_data/1CSV_Voice.csv')
for num in np.arange(2, 6, 1):
    next_phone = pd.read_csv('./phone_data/'+str(num)+'CSV_Voice.csv')
    full_phone = pd.concat([full_phone, next_phone], sort=False)
full_phone['Direction'] = full_phone['Destination'].str.upper() != 'INCOMING'
full_phone['Direction'].replace({True: 'Outgoing', False: 'Incoming'}, inplace=True)
full_phone.drop(['Destination', 'Number', 'Call Type', 'Time', 'Charge'], axis=1, inplace=True)
    
full_text = pd.read_csv('./phone_data/1CSV_Messages.csv')
for num in np.arange(2, 6, 1):
    next_text = pd.read_csv('./phone_data/'+str(num)+'CSV_Messages.csv')
    full_text = pd.concat([full_text, next_text], sort=False)
full_text.drop(['Destination', 'Number', 'Charge', 'Time', 'Type'], axis=1, inplace=True)

# Select January 1 to May 1 only
full_phone['Date'] = pd.to_datetime(full_phone['Date'])
full_text['Date'] = pd.to_datetime(full_text['Date'])

full_phone = full_phone.loc[(full_phone['Date'] >= '2020-01-01') & (full_phone['Date'] < '2020-05-01')]
full_text = full_text.loc[(full_text['Date'] >= '2020-01-01') & (full_text['Date'] < '2020-05-01')]

print('Dataset length:', len(full_phone), '&', len(full_text))
display(full_phone.head())

full_phone.to_csv('./phone_data/Eleanor_phone_records.csv', index=False)
full_text.to_csv('./phone_data/Eleanor_text_records.csv', index=False)
```

```python
# Clean individual phone calls and texts into daily totals

phone = pd.read_csv('./phone_data/Eleanor_phone_records.csv')
text = pd.read_csv('./phone_data/Eleanor_text_records.csv')
text['Date'] = pd.to_datetime(text['Date'])
phone['Date'] = pd.to_datetime(phone['Date'])

phone['Minutes'] = phone['Minutes'].str.replace(' Min', '').astype(int)
phone = phone.groupby(['Date', 'Direction']).sum()
phone.reset_index(inplace=True)
display(phone.tail())

text['Count'] = 1
text = text.groupby(['Date', 'Direction']).count()
text.reset_index(inplace=True)
display(text.tail())

print('Data length now:', len(text), len(phone))
print('Days in Jan - Apr:', 31 + 29 + 31 + 30)
print('Check if ok? Should not be more than days:', len(text)/2, len(phone)/2)
```

```python
# Fill missing dates with 0 values, to make it easier for plotting

def fill_missing_split(df):
    df_in = df[df['Direction']=='Incoming']
    df_out = df[df['Direction']=='Outgoing']
    assert(len(df_in) + len(df_out) == len(df))

    df_in.index = df_in['Date']
    df_in = df_in.reindex(pd.date_range(start='01-01-2020', end='04-30-2020'), fill_value = 0)
    df_in.drop('Date', axis=1, inplace=True)
    df_in.reset_index(inplace=True)
    df_in.rename(columns={'index':'Date'}, inplace=True)
    df_in['Day'] = df_in['Date'] - pd.to_datetime('2020-01-01 00:00:00')
    df_in['Day'] = pd.to_timedelta(df_in['Day'], unit='d').dt.days
    assert((list(set(df_in['Direction'].values)) == [0, 'Incoming']) | \
           (list(set(df_in['Direction'].values)) == ['Incoming']))
    df_in['Direction'] = 'Incoming'
    df_in.sort_values(by=['Day'], axis=0, ascending=True, inplace=True)
    
    df_out.index = df_out['Date']
    df_out = df_out.reindex(pd.date_range(start='01-01-2020', end='04-30-2020'), fill_value = 0)
    df_out.drop('Date', axis=1, inplace=True)
    df_out.reset_index(inplace=True)
    df_out.rename(columns={'index':'Date'}, inplace=True)
    df_out['Day'] = df_out['Date'] - pd.to_datetime('2020-01-01 00:00:00')
    df_out['Day'] = pd.to_timedelta(df_out['Day'], unit='d').dt.days
    assert((list(set(df_out['Direction'].values)) == [0, 'Outgoing']) | \
           (list(set(df_out['Direction'].values)) == ['Outgoing']))
    df_out['Direction'] = 'Outgoing'
    df_out.sort_values(by=['Day'], axis=0, ascending=True, inplace=True)

    assert(len(df_in) == len(df_out) == 121)
    return df_in, df_out

phone_in, phone_out = fill_missing_split(phone)
text_in, text_out = fill_missing_split(text)

display(phone_in.head())
display(text_out.head())
```

```python
fig = plt.figure(figsize = (5, 3))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

pink1 = '#f87074'
pink2 = '#ff9396'
blue1 = '#75bdd2'
blue2 = '#99e1e8'

x1, y1 = phone_in['Day'].tolist(), phone_in['Minutes'].tolist()
x2, y2 = phone_out['Day'].tolist(), phone_out['Minutes'].tolist()
limit = [x+y for x, y in zip(y1, y2)]
print('Max min phone:', max(limit), min(limit))
ax1.bar(x1, y1, width=1, color=pink2, bottom=y2, clip_on=False, align='edge')
ax1.bar(x2, y2, width=1, color=pink1, clip_on=False, align='edge')
ax1.set_ylim(0, 180)

x1, y1 = text_in['Day'].tolist(), text_in['Count'].tolist()
x2, y2 = text_out['Day'].tolist(), text_out['Count'].tolist()
limit = [x+y for x, y in zip(y1, y2)]
print('Max min text:', max(limit), min(limit))
ax2.bar(x1, y1, width=1, color=blue2, bottom=y2, clip_on=False, align='edge')
ax2.bar(x2, y2, width=1, color=blue1, clip_on=False, align='edge')
ax2.set_ylim(180, 0)

ax2.plot([0, 121], [0, 0], clip_on=False, zorder=100, color='k', lw=1.4) # x axis
ax1.set_yticks([60, 120, 180])
ax1.set_yticklabels([1, 2, '3 hours'])
ax2.set_yticks([0, 50, 100])
ax2.set_yticklabels([0, 50, '100 texts'])

# WA stay home order date
mark_date = pd.to_datetime('2020-03-23 00:00:00') - pd.to_datetime('2020-01-01 00:00:00')
mark_date = pd.to_timedelta(mark_date, unit='d').days

for ax in [ax1, ax2]:
    ax.plot([mark_date, mark_date], [0, 180], color='k', lw=1.4)
    ax.set_xlim(0, 121)
    ax.yaxis.set_label_position("right")
    ax.yaxis.tick_right()
    ax.axes.get_xaxis().set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    
months = [0, 31, 29, 31, 30]
descs = ['JAN', 'FEB', 'MAR', 'APR']
months = np.cumsum(months)
for month, desc in zip(months, descs): 
    ax2.plot([month, month], [0, 10], zorder=100, color='k', lw=1.4, clip_on=False)
    ax2.text(month+1, 5, desc, verticalalignment='top', horizontalalignment='left')
    
plt.subplots_adjust(top=1, bottom=0, right=1, left=0, hspace=0, wspace=0)
plt.margins(0,0)
plt.show()

pp = pdf.PdfPages('./phone_data/figure.pdf', keep_empty=False)
pp.savefig(fig)
pp.close()
```
